```python
def build_initial_ionpair(dt,eq_cutoff,eq_temp,eq_steps):

    import pandas as pd
    
    import time 
    init_time = time.time()
    
    dt = dt

    import MDAnalysis as mda
    from nglview.datafiles import PDB, XTC


    #混合溶液を作成
    import mdapackmol
    import numpy as np
    from ase import units

    # load individual molecule files
    mol1 = mda.Universe('anion.gro')
    mol2 = mda.Universe('cation.gro')
    num_mols1 = 1
    num_mols2 = 1

    mw_mol1 = np.sum(mol1.atoms.masses)
    mw_mol2 = np.sum(mol2.atoms.masses)
    total_weight = num_mols1 * mw_mol1 

    # Determine side length of a box with the density of mixture 
    L = 50.0 # Ang. unit 

    system = mdapackmol.packmol(
    [ mdapackmol.PackmolStructure(
    mol1, number=1,
    instructions=['inside box 0. 0. 0. '+str(L)+"  "+str(L)+"  "+str(L)]),
    mdapackmol.PackmolStructure(
    mol2, number=1,
    instructions=['inside box 0. 0. 0. '+str(L)+"  "+str(L)+"  "+str(L)]),
    ])

    system.atoms.write('mixture.gro')

    import os 
    os.environ['GMX_MAXBACKUP'] = '-1'

    # for gromacs-5 or later 
    commands = "gmx editconf -f mixture.gro  -box "+ str(L/10.0)+"  "+str(L/10.0)+"  "+str(L/10.0) + "  " +" -o init.gro"
    #commands = "editconf -f mixture.gro  -box "+ str(L/10.0)+"  "+str(L/10.0)+"  "+str(L/10.0) + "  " +" -o init.gro"
    
    import subprocess
    from subprocess import PIPE

    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout
    #print('STDOUT: {}'.format(output))

    #make top file for GAFF

    top_file = "system.top"
    mol_name1 = "anion"
    mol_name2 = "cation"

    lines = [
        "; input_GMX.top created by acpype (v: 2020-07-25T09:06:13CEST) on Fri Jul 31 07:59:08 2020",
        ";by acpype (v: 2020-07-25T09:06:13CEST) on Fri Jul 31 07:59:08 2020",
        "   ",
        "[ defaults ]",
        "; nbfunc        comb-rule       gen-pairs       fudgeLJ fudgeQQ",
        "1               2               yes             0.5     0.8333",
        "    ",
        "; Include input.itp topology", 
        "#include \"{}.itp\"".format(mol_name1),
        "#include \"{}.itp\"".format(mol_name2),
        "    ",
        "[ system ]",
        "input",
        "     ",
        "[ molecules ]",
        "; Compound        nmols" ,
        mol_name1 + "          {} ".format(num_mols1), 
        mol_name2 + "          {} ".format(num_mols2)
    ]
        
    with open(top_file, mode='w') as f:
        f.write('\n'.join(lines))

    import sys
    import mdtraj

    # OpenMM Imports
    import simtk.openmm as mm
    import simtk.openmm.app as app
    from openmmtools import integrators

    # ParmEd Imports
    from parmed import load_file
    from parmed.openmm.reporters import NetCDFReporter
    from parmed import unit as u

    # Load in a gromacs system
    from simtk.openmm import app
    gro = app.GromacsGroFile('init.gro')
    top = app.GromacsTopFile('system.top', periodicBoxVectors=gro.getPeriodicBoxVectors(), includeDir="./")

    # Equilibration
    print('Creating OpenMM System')

    system = top.createSystem(nonbondedMethod=app.PME,
                  nonbondedCutoff=eq_cutoff*u.angstroms,
                  #constraints=app.HBonds,
                  removeCMMotion=True,
    )

    # Create the integrator to do Langevin dynamics
    integrator = mm.LangevinIntegrator(
                            eq_temp*u.kelvin,       # Temperature of heat bath
                            1.0/u.picoseconds,  # Friction coefficient
                            dt*u.femtoseconds, # Time step
    )

    timestep = dt * u.femtoseconds
    collision_rate = 1.0 / u.picoseconds
    temperature = eq_temp * u.kelvin
    integrator = integrators.AndersenVelocityVerletIntegrator(temperature, collision_rate, timestep)

    # Define the platform to use; CUDA, OpenCL, CPU, or Reference. Or do not specify
    # the platform to use the default (fastest) platform
    platform = mm.Platform.getPlatformByName('CPU')
    #prop = dict(CudaPrecision='mixed') # Use mixed single/double precision

    # Create the Simulation object
    #sim = app.Simulation(top.topology, system, integrator, platform, prop)
    sim = app.Simulation(top.topology, system, integrator, platform)

    # Set the particle positions
    sim.context.setPositions(gro.positions)

    # Minimize the energy
    print('Minimizing energy')
    sim.minimizeEnergy(maxIterations=10000)

    #Relax the geometry 
    print('Running dynamics :Equilibration')
    sim.context.setVelocitiesToTemperature(eq_temp*u.kelvin) 
    sim.reporters.append(
        mdtraj.reporters.HDF5Reporter('eq_out.h5', 100, coordinates=True, time=True, 
                                      cell=True, potentialEnergy=True, temperature=True))
    sim.reporters.append(
        app.StateDataReporter("eq_ene.csv", 100, step=True, potentialEnergy=True,
                              kineticEnergy=True, temperature=True))
    sim.step(eq_steps)

    del sim # Make sure to close all files
    
    print("elapsed time= {} sec.".format(time.time()-init_time))
```

```python
def tail(fn, n):
    # ファイルを開いてすべての行をリストで取得する
    with open(fn, 'r') as f:
        f.readline()
        lines = f.readlines()

    # 文字列を配列にしてから返す. ついでにstr->floatに型変換する
    return [line.strip().split() for line in lines[-n:]]
```

```python
#make mdp file for energy minimization
def make_mdp_em():

    mdp_file = "em.mdp"

    lines = [
    "; VARIOUS PREPROCESSING OPTIONS",
    ";title                    = Yo",
    ";cpp                      = /usr/bin/cpp",
    "include                  =", 
    "define                   =", 
    "    ",
    "; RUN CONTROL PARAMETERS",
    "integrator               = steep",
    "nsteps                   = 1000000",
    "emtol                    = 10",
    "emstep                   = 0.1",
    "nstlist                  = 1",
    "cutoff-scheme            = verlet",
    "vdw-type                 = cut-off",
    "rlist                    = 2.0",
    "rvdw                     = 2.0",
    "rcoulomb                 = 2.0",
    ]

    with open(mdp_file, mode='w') as f:
        f.write('\n'.join(lines))
```

```python
#Equilibration of all the Replica 

#make mdp file for NVT run
def make_mdp_nvt(temp,steps,dt):

    temperature      = temp
    simulation_steps = steps 
    time_step        = dt/1000.0  # ps
    
    mdp_file = "nvt{}.mdp".format(temp)

    lines = [
    "; RUN CONTROL PARAMETERS",
    "constraints               = all-bonds",
    "integrator               = sd",
    "nsteps                   = {}".format(simulation_steps),
    "dt                       = {}".format(time_step),
    "nstlist                  = 10",
    "rlist                    = 2",
    "rvdw                     = 2",
    "rcoulomb                 = 2",
    #"cutoff-scheme            = group",
    #"ns_type                  = simple",
    "coulombtype              = Cut-off",
    "vdw-type                 = Cut-off",        
    "tcoupl                   = v-rescale",
    "tc-grps                  = system",
    "tau-t                    = 2.0",
    "ref-t                    = {}".format(temperature),
    "Pcoupl                   = no",
    "nstenergy                = 1000",
    "nstxout                  = 1000",
    #"pbc                      = no",
    ]

    with open(mdp_file, mode='w') as f:
        f.write('\n'.join(lines))
```

```python
#make mdp file for NVT run for all replicas
def make_mdp_nvt_replica(temp,steps,dt):

    temperature      = temp
    simulation_steps = steps 
    time_step        = dt/1000.0  # ps
    
    mdp_file = "nvt{}.mdp".format(temp)

    lines = [
    "; RUN CONTROL PARAMETERS",
    "constraints               = all-bonds",
    "integrator               = sd",
    "nsteps                   = {}".format(simulation_steps),
    "dt                       = {}".format(time_step),
    "nstlist                  = 10",
    "rlist                    = 2.0",
    "rvdw                     = 2.0",
    "rcoulomb                 = 2.0",
    #"cutoff-scheme            = group",
    #"ns_type                  = simple",
    "coulombtype              = Cut-off",
    "vdw-type                 = Cut-off",        
    "tcoupl                   = v-rescale",
    "tc-grps                  = system",
    "tau-t                    = 2.0",
    "ref-t                    = {}".format(temperature),
    "Pcoupl                   = no",
    "nstenergy                = 1000",
    "nstxout                  = 1000",
    #"pbc                      = no",
    ]

    with open(mdp_file, mode='w') as f:
        f.write('\n'.join(lines))
```

```python
#make mdp file for NVT run for all replicas
def make_mdp_nvt_replica_FEP(temp,steps,dt,lambdas):
    
    l=str(lambdas)
    l=l.replace(","," ")
    l=l.replace("["," ")
    l=l.replace("]"," ")

    temperature      = temp
    simulation_steps = steps 
    time_step        = dt/1000.0  # ps
    
    mdp_file = "nvt{}.mdp".format(temp)

    lines = [
    "; RUN CONTROL PARAMETERS",
    "constraints               = all-bonds",
    "integrator               = sd",
    "nsteps                   = {}".format(simulation_steps),
    "dt                       = {}".format(time_step),
    "nstlist                  = 10",
    "rlist                    = 2.0",
    "rvdw                     = 2.0",
    "rcoulomb                 = 2.0",
    #"cutoff-scheme            = group",
    #"ns_type                  = simple",
    "coulombtype              = Cut-off",
    "vdw-type                 = Cut-off",        
    "tcoupl                   = v-rescale",
    "tc-grps                  = system",
    "tau-t                    = 2.0",
    "ref-t                    = {}".format(temperature),
    "Pcoupl                   = no",
    "nstenergy                = 1000",
    "nstxout                  = 1000",
    #"pbc                      = no",
    "; and set the free energy parameters",
    "free-energy              = yes", 
    "couple-moltype           = input",
    "; these 'soft-core' parameters make sure we never get overlapping",
    "; charges as lambda goes to 0",
    "sc-power                 = 1",
    "sc-sigma                 = 0.3",
    "sc-alpha                 = 1.0",
    "; we still want the molecule to interact with itself at lambda=0",
    "couple-intramol          = no",
    "couple-lambda1           = vdwq",
    "couple-lambda0           = none",
    "; These are the lambda states at which we simulate",
    "; for separate LJ and Coulomb decoupling, use",
    "fep-lambdas              = {} ".format(l),
    "init-lambda-state        = 0",
    ]

    with open(mdp_file, mode='w') as f:
        f.write('\n'.join(lines))
```

```python
def make_atomtypes(itps):
    
    atoms_data = []
    
    for itp_name in itps :
        
        # itp-file よりatomtypesを抽出して削除する
        f = open('{}.itp'.format(str(itp_name)), 'r')
        flag=0 
        mod_itp = []
        while True:
          data = f.readline()
          if data == '':
            break
          else:
            if data =="[ atomtypes ]\n":
                flag=1
            if data =="\n" and flag==1:
                flag=0
            if flag==1:
                atoms_data.append(data)
                data=""
            else :
                mod_itp.append(data)
        f.close()

        #編集結果をitpファイルに書き込む
        f = open('{}.itp'.format(str(itp_name)), 'w')
        f.writelines(mod_itp)
        f.close()
    
    #atomtypesを別ファイルに書き出して終了
    f = open('atomlist.itp', 'w')
    f.writelines(sorted(set(atoms_data), key=atoms_data.index))
    f.close()
```

```python
#make mdp file for NVT run
def make_mdp_nvt(temp,steps,dt,cutoff,coulombtype,velocity,intg):

    temperature      = temp
    simulation_steps = steps 
    time_step        = dt/1000.0  # ps
    cutoff_radius    = cutoff
    intg_type        = intg
    
    mdp_file = "run.mdp"

    lines = [
    "; VARIOUS PREPROCESSING OPTIONS",
    ";title                    = Yo",
    ";cpp                      = /usr/bin/cpp",
    "include                  =", 
    "define                   =", 
    "    ",
    "; RUN CONTROL PARAMETERS",
    "constraints              = none",
    "integrator               = {}".format(intg_type),
    "nsteps                   = {}".format(simulation_steps),
    "dt                       = {}".format(time_step),
    "nstlist                  = 50",
    "rlist                    = {}".format(cutoff_radius),
    "rvdw                     = {}".format(cutoff_radius),
    "rcoulomb                 = {}".format(cutoff_radius),
    "coulombtype              = {}".format(coulombtype),
    "cutoff-scheme            = verlet",
    "vdw-type                 = cut-off",
    "tcoupl                   = v-rescale",
    "tc-grps                  = system",
    "tau-t                    = 2.0",
    "ref-t                    = {}".format(temperature),
    "Pcoupl                   = no",
    "nstenergy                = 100",
    "nstxout                  = 1000",
    "gen-vel                  = yes" ,
    "gen-temp                 = {}".format(temperature),
    "gen-seed                 = -1 ",
    "deform                   = {0} {0} {0} 0 0 0".format(-velocity)    
    ]

    with open(mdp_file, mode='w') as f:
        f.write('\n'.join(lines))
```

```python
#make_initial_structure 

def calc_rdf(inp_file,top_file,cond):
    
    import shutil
    import os 
    import subprocess
    from subprocess import PIPE
    import re 
    import numpy as np
    
    
    os.environ['GMX_MAXBACKUP'] = '-1'
    !. /home/yamazaki/usr/local/gromacs/bin/GMXRC
    
    
    gromacs_home = "/home/yamazaki/usr/local/gromacs/bin/"
    
    #calculation
    dt = 1 
    
    !echo PAG > ./inputs/anal.txt 
    !echo POL >> ./inputs/anal.txt 
    !echo  >> ./inputs/anal.txt 
    
    temperature = cond[0]
    pressure    = cond[1]
    md_steps    = cond[2]
    coul_type   = cond[3]
    cutoff      = cond[4]
    speed       = cond[5]
    intg_type   = cond[6]
    #print("cond=",cond)
    if pressure < 0 :
        make_mdp_nvt(temperature,md_steps,dt,cutoff,coul_type,speed,intg_type)
    else :
        make_mdp_run(temperature,pressure,md_steps,dt,cutoff,coul_type,speed,intg_type)

    #grompp
    os.environ['OMP_NUM_THREADS'] = '1'    
    commands = gromacs_home+"gmx_mpi grompp -f run.mdp -p {1}.top -c {0}.gro -o run.tpr -maxwarn 10 ".format(inp_file,top_file)
    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout

    #mdrun
    os.environ['OMP_NUM_THREADS'] = '6'  
    commands = gromacs_home+"gmx_mpi mdrun -s run.tpr -o run.trr -e run.edr -c run.gro -nb gpu -bonded gpu"
    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout
    
    #analysis
    os.environ['OMP_NUM_THREADS'] = '1'  
    commands = gromacs_home+"gmx_mpi rdf -s run.tpr -f run.trr -b {0} -o rdf.xvg -selrpos mol_com -seltype mol_com -bin 0.01  < ./inputs/anal.txt".format(min(int(md_steps/1000/2),1000))
    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout

    #read xvg file 
    path ="rdf.xvg"
    with open(path) as f:
        l_strip = [s.strip() for s in f.readlines() if ("#" not in s) and ("@" not in s) ]
        rdf   = [float(re.split(" +",line)[1]) for line in l_strip]
        r_list= [float(re.split(" +",line)[0]) for line in l_strip]
     
    return [r_list,rdf]
```
