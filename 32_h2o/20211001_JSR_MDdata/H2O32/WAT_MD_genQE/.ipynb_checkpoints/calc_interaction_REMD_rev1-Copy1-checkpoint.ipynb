```python
def build_initial_ionpair(dt,eq_cutoff,eq_temp,eq_steps):

    import pandas as pd
    
    import time 
    init_time = time.time()
    
    dt = dt

    import MDAnalysis as mda
    from nglview.datafiles import PDB, XTC


    #混合溶液を作成
    import mdapackmol
    import numpy as np
    from ase import units

    # load individual molecule files
    mol1 = mda.Universe('anion.gro')
    mol2 = mda.Universe('cation.gro')
    num_mols1 = 1
    num_mols2 = 1

    mw_mol1 = np.sum(mol1.atoms.masses)
    mw_mol2 = np.sum(mol2.atoms.masses)
    total_weight = num_mols1 * mw_mol1 

    # Determine side length of a box with the density of mixture 
    L = 50.0 # Ang. unit 

    system = mdapackmol.packmol(
    [ mdapackmol.PackmolStructure(
    mol1, number=1,
    instructions=['inside box 0. 0. 0. '+str(L)+"  "+str(L)+"  "+str(L)]),
    mdapackmol.PackmolStructure(
    mol2, number=1,
    instructions=['inside box 0. 0. 0. '+str(L)+"  "+str(L)+"  "+str(L)]),
    ])

    system.atoms.write('mixture.gro')

    import os 
    os.environ['GMX_MAXBACKUP'] = '-1'

    # for gromacs-5 or later 
    commands = "gmx editconf -f mixture.gro  -box "+ str(L/10.0)+"  "+str(L/10.0)+"  "+str(L/10.0) + "  " +" -o init.gro"
    #commands = "editconf -f mixture.gro  -box "+ str(L/10.0)+"  "+str(L/10.0)+"  "+str(L/10.0) + "  " +" -o init.gro"
    
    import subprocess
    from subprocess import PIPE

    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout
    #print('STDOUT: {}'.format(output))

    #make top file for GAFF

    top_file = "system.top"
    mol_name1 = "anion"
    mol_name2 = "cation"

    lines = [
        "; input_GMX.top created by acpype (v: 2020-07-25T09:06:13CEST) on Fri Jul 31 07:59:08 2020",
        ";by acpype (v: 2020-07-25T09:06:13CEST) on Fri Jul 31 07:59:08 2020",
        "   ",
        "[ defaults ]",
        "; nbfunc        comb-rule       gen-pairs       fudgeLJ fudgeQQ",
        "1               2               yes             0.5     0.8333",
        "    ",
        "; Include input.itp topology", 
        "#include \"{}.itp\"".format(mol_name1),
        "#include \"{}.itp\"".format(mol_name2),
        "    ",
        "[ system ]",
        "input",
        "     ",
        "[ molecules ]",
        "; Compound        nmols" ,
        mol_name1 + "          {} ".format(num_mols1), 
        mol_name2 + "          {} ".format(num_mols2)
    ]
        
    with open(top_file, mode='w') as f:
        f.write('\n'.join(lines))

    import sys
    import mdtraj

    # OpenMM Imports
    import simtk.openmm as mm
    import simtk.openmm.app as app
    from openmmtools import integrators

    # ParmEd Imports
    from parmed import load_file
    from parmed.openmm.reporters import NetCDFReporter
    from parmed import unit as u

    # Load in a gromacs system
    from simtk.openmm import app
    gro = app.GromacsGroFile('init.gro')
    top = app.GromacsTopFile('system.top', periodicBoxVectors=gro.getPeriodicBoxVectors(), includeDir="./")

    # Equilibration
    print('Creating OpenMM System')

    system = top.createSystem(nonbondedMethod=app.PME,
                  nonbondedCutoff=eq_cutoff*u.angstroms,
                  #constraints=app.HBonds,
                  removeCMMotion=True,
    )

    # Create the integrator to do Langevin dynamics
    integrator = mm.LangevinIntegrator(
                            eq_temp*u.kelvin,       # Temperature of heat bath
                            1.0/u.picoseconds,  # Friction coefficient
                            dt*u.femtoseconds, # Time step
    )

    timestep = dt * u.femtoseconds
    collision_rate = 1.0 / u.picoseconds
    temperature = eq_temp * u.kelvin
    integrator = integrators.AndersenVelocityVerletIntegrator(temperature, collision_rate, timestep)

    # Define the platform to use; CUDA, OpenCL, CPU, or Reference. Or do not specify
    # the platform to use the default (fastest) platform
    platform = mm.Platform.getPlatformByName('CPU')
    #prop = dict(CudaPrecision='mixed') # Use mixed single/double precision

    # Create the Simulation object
    #sim = app.Simulation(top.topology, system, integrator, platform, prop)
    sim = app.Simulation(top.topology, system, integrator, platform)

    # Set the particle positions
    sim.context.setPositions(gro.positions)

    # Minimize the energy
    print('Minimizing energy')
    sim.minimizeEnergy(maxIterations=10000)

    #Relax the geometry 
    print('Running dynamics :Equilibration')
    sim.context.setVelocitiesToTemperature(eq_temp*u.kelvin) 
    sim.reporters.append(
        mdtraj.reporters.HDF5Reporter('eq_out.h5', 100, coordinates=True, time=True, 
                                      cell=True, potentialEnergy=True, temperature=True))
    sim.reporters.append(
        app.StateDataReporter("eq_ene.csv", 100, step=True, potentialEnergy=True,
                              kineticEnergy=True, temperature=True))
    sim.step(eq_steps)

    del sim # Make sure to close all files
    
    print("elapsed time= {} sec.".format(time.time()-init_time))
```

```python
def tail(fn, n):
    # ファイルを開いてすべての行をリストで取得する
    with open(fn, 'r') as f:
        f.readline()
        lines = f.readlines()

    # 文字列を配列にしてから返す. ついでにstr->floatに型変換する
    return [line.strip().split() for line in lines[-n:]]
```

```python
def build_gas_cell(name,cell_length,n_cat,n_ani,n_poly):

    import pandas as pd
    import shutil
    
    #構造可視化
    import MDAnalysis as mda

    #混合溶液を作成
    import mdapackmol
    import numpy as np
    from ase import units

    L=cell_length
    num_mols1 = n_cat
    num_mols2 = n_ani
    num_mols3 = n_poly
    mol_name1 = "anion"
    mol_name2 = "cation"
    mol_name3 = "input"
    itps = []
    
    copy = './anion.itp'
    src = "./output/"+str(name)+'_anion.itp'
    shutil.copyfile(src,copy)
    copy = './cation.itp'
    src = "./output/"+str(name)+'_cation.itp'
    shutil.copyfile(src,copy)
    copy = './pag.gro'
    src = "./output/"+str(name)+'_pair.gro'
    shutil.copyfile(src,copy)        
    copy = './input1.itp'
    src = "./output/"+str(name)+'_poly.itp'
    shutil.copyfile(src,copy)
    copy = './input1.gro'
    src = "./output/"+str(name)+'_poly.gro'
    shutil.copyfile(src,copy)

    #各itpsよりatomtypesを抽出、編集する
    itps.append("anion")
    itps.append("cation")
    itps.append("input1")
    make_atomtypes(itps)
    
    if (n_ani>0) and (n_cat>0) and (n_poly>0):
    
        # load individual molecule files
        mol1 = mda.Universe('pag.gro')
        mol2 = mda.Universe('input1.gro')

        system = mdapackmol.packmol(
        [ mdapackmol.PackmolStructure(
        mol1, number=num_mols1,
        instructions=["inside box "+str(0)+" "+str(0)+" "+str(0)+" "+str(L)+" "+str(L)+" "+str(L)]),
        mdapackmol.PackmolStructure(
        mol2, number=num_mols2,
        instructions=["inside box "+str(0)+" "+str(0)+" "+str(0)+" "+str(L)+" "+str(L)+" "+str(L)]),
        ])

        system.atoms.write('gas_cell.gro')
         
    if (n_ani>0) and (n_cat>0) and (n_poly==0):
    
        # load individual molecule files
        mol1 = mda.Universe('pag.gro')

        system = mdapackmol.packmol(
        [ mdapackmol.PackmolStructure(
        mol1, number=num_mols1,
        instructions=["inside box "+str(0)+" "+str(0)+" "+str(0)+" "+str(L)+" "+str(L)+" "+str(L)]),
        ])

        system.atoms.write('gas_cell.gro')
    
    if (n_ani==0) and (n_cat==0) and (n_poly>0):
    
        # load individual molecule files
        mol2 = mda.Universe('input1.gro')

        system = mdapackmol.packmol(
        [ mdapackmol.PackmolStructure(
        mol2, number=num_mols3,
        instructions=["inside box "+str(0)+" "+str(0)+" "+str(0)+" "+str(L)+" "+str(L)+" "+str(L)]),
        ])

        system.atoms.write('gas_cell.gro')

    import os 
    os.environ['GMX_MAXBACKUP'] = '-1'
    
    gromacs_home = "/home/yam13524/usr/local/gromacs2018/bin/"
    !export OMP_NUM_THREADS=1   
    commands = gromacs_home + "gmx_mpi editconf -f gas_cell.gro  -box "+ str(L/10.0)+"  "+str(L/10.0)+"  "+str(L/10.0) + "  " +" -o init1.gro"

    import subprocess
    from subprocess import PIPE

    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout
    #print('STDOUT: {}'.format(output))

    #make top file for GAFF

    top_file = "system_vac.top"

    lines = [
        "; input_GMX.top created by acpype (v: 2020-07-25T09:06:13CEST) on Fri Jul 31 07:59:08 2020",
        ";by acpype (v: 2020-07-25T09:06:13CEST) on Fri Jul 31 07:59:08 2020",
        "   ",
        "[ defaults ]",
        "; nbfunc        comb-rule       gen-pairs       fudgeLJ fudgeQQ",
        "1               2               yes             0.5     0.8333",
        "    ",
        "; Include input.itp topology", 
        "#include \"atomlist.itp\"",
        "#include \"anion.itp\"",
        "#include \"cation.itp\"",
        "#include \"input1.itp\"",
        
        "    ",
        "[ system ]",
        "input",
        "     ",
        "[ molecules ]",
        "; Compound        nmols" ,
        mol_name1 + "          " + str(num_mols1),
        mol_name2 + "          " + str(num_mols2),
        mol_name3 + "          " + str(num_mols3),]
                
    with open(top_file, mode='w') as f:
        f.write('\n'.join(lines))
    
    make_mdp_em()
                
    #grompp
    os.environ['OMP_NUM_THREADS'] = '1'    
    commands = gromacs_home+"gmx_mpi grompp -f em.mdp -p system_vac.top -c init1.gro -o em.tpr -maxwarn 10 "
    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout

    #mdrun
    os.environ['OMP_NUM_THREADS'] = '1' 
    commands = gromacs_home+"gmx_mpi mdrun -s em.tpr -o em.trr -e em.edr -c init_gas.gro -nb cpu"
    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout
    
    line = np.array([float(p) for p in tail("{}.gro".format("init_gas"),1)[0]]) 
    return [line[0],line[1],line[2],num_mols1]
```

```python
#make mdp file for energy minimization
def make_mdp_em():

    mdp_file = "em.mdp"

    lines = [
    "; VARIOUS PREPROCESSING OPTIONS",
    ";title                    = Yo",
    ";cpp                      = /usr/bin/cpp",
    "include                  =", 
    "define                   =", 
    "    ",
    "; RUN CONTROL PARAMETERS",
    "integrator               = steep",
    "nsteps                   = 1000000",
    "emtol                    = 10",
    "emstep                   = 0.1",
    "nstlist                  = 1",
    "cutoff-scheme            = verlet",
    "vdw-type                 = cut-off",
    "rlist                    = 2.0",
    "rvdw                     = 2.0",
    "rcoulomb                 = 2.0",
    ]

    with open(mdp_file, mode='w') as f:
        f.write('\n'.join(lines))
```

```python
def make_atomtypes(itps):
    
    atoms_data = []
    
    for itp_name in itps :
        
        # itp-file よりatomtypesを抽出して削除する
        f = open('{}.itp'.format(str(itp_name)), 'r')
        flag=0 
        mod_itp = []
        while True:
          data = f.readline()
          if data == '':
            break
          else:
            if data =="[ atomtypes ]\n":
                flag=1
            if data =="\n" and flag==1:
                flag=0
            if flag==1:
                atoms_data.append(data)
                data=""
            else :
                mod_itp.append(data)
        f.close()

        #編集結果をitpファイルに書き込む
        f = open('{}.itp'.format(str(itp_name)), 'w')
        f.writelines(mod_itp)
        f.close()
    
    #atomtypesを別ファイルに書き出して終了
    f = open('atomlist.itp', 'w')
    f.writelines(sorted(set(atoms_data), key=atoms_data.index))
    f.close()
```

```python
def pre_REMD(dt,temps,total_steps):
    #Equilibration of all the Replica 
    import shutil
    import os 
    import subprocess
    from subprocess import PIPE

    os.environ['GMX_MAXBACKUP'] = '-1'
    !. /home/yam13524/usr/local/gromacs2018/bin/GMXRC
    gromacs_home = "/home/yam13524/usr/local/gromacs2018/bin/"

    total_steps = total_steps
    dt = dt 
    
    for temp in temps : 
        make_mdp_nvt_replica(temp,total_steps,dt)

        #grompp
        os.environ['OMP_NUM_THREADS'] = '1'
        commands = gromacs_home+"gmx_mpi grompp -f nvt{}.mdp -p system_vac.top -c min.gro -o nvt.tpr -maxwarn 10 ".format(temp)
        proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
        output = proc.stdout

        #mdrun
        os.environ['OMP_NUM_THREADS'] = '1' 
        commands = gromacs_home+"gmx_mpi mdrun -s nvt.tpr -o nvt{0}.trr -e nvt{0}.edr -c nvt{0}.gro -nb cpu ".format(temp)
        proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
        output = proc.stdout   
```

```python
def REMD(dt,temps,total_steps):
    
    import shutil
    import os 
    import subprocess
    from subprocess import PIPE
    
    #REMD Simulation
    os.environ['GMX_MAXBACKUP'] = '-1'
    !. /home/yam13524/usr/local/gromacs2018/bin/GMXRC
    gromacs_home = "/home/yam13524/usr/local/gromacs2018/bin/"
    os.environ['OMP_NUM_THREADS'] = '1'
    total_steps = total_steps
    dt = dt
    for i,temp in enumerate(temps):
        #make_mdp_nvt(temp,total_steps,dt)
        make_mdp_nvt_replica_woFEP(temp,total_steps,dt)
        #grompp   
        commands = gromacs_home+"gmx_mpi grompp -f nvt{0}.mdp -p system_vac.top -c nvt{0}.gro -o remd{1}.tpr -maxwarn 10 ".format(temp,i)
        proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
        output = proc.stdout

    #mdrun
    os.environ['OMP_NUM_THREADS'] = '1' 
    commands = "mpirun -np 10 " + gromacs_home + "gmx_mpi mdrun  -deffnm remd -multi 10 -replex 1000 -nb cpu "
    proc = subprocess.run(commands, shell=True, stdout=PIPE, stderr=PIPE,encoding='utf-8')
    output = proc.stdout
```
