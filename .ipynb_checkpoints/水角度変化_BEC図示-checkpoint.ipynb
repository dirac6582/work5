#  H，OのBECを読み出して単純に図示する．


## ファイル読み込み→ 11成分のみ取り出し→listにappend→ 図示の順番が良さそう．

```python
import numpy as np
```

```python
# 有効電荷の取得ができるかのテスト
ZO=np.loadtxt("./water_anglechange/result/bec_O_100.txt")
ZH1=np.loadtxt("./water_anglechange/result/bec_H1_100.txt")
ZH2=np.loadtxt("./water_anglechange/result/bec_H2_100.txt")

print(ZO)
print(ZH1)
print(ZH2)
```

```python
#===============================
# O原子のBEC対角成分
#===============================


bec11_list=[]
bec22_list=[]
bec33_list=[]

i_array=np.arange(80,130)

for i in range(80,130):
    print(i)
    # zero padding
    #count_zero = str(i).zfill(4)
    #print(count_zero)
    #BECarray[i,:,:,:]=np.load("WAT32_BEC/"+count_zero+".npy")
    ZO=np.loadtxt("./water_anglechange/result/bec_O_"+str(i)+".txt")
    ## 対角成分の取り出し
    bec11_list.append(ZO[0][0])
    bec22_list.append(ZO[1][1])
    bec33_list.append(ZO[2][2])
    print(ZO)
    ##print(BECarray[i,:,:,:])
```

```python
# np.arrayに変更
bec11_array=np.array(bec11_list)
bec22_array=np.array(bec22_list)
bec33_array=np.array(bec33_list)
```

```python
# 得られたdetたちを図示する．
import numpy as np
import matplotlib.pyplot as plt

# ここにグラフを描画する処理を記載する．
print(i_array)

plt.plot(i_array,bec11_array, label="ZO11")
plt.plot(i_array,bec22_array, label="ZO22")
plt.plot(i_array,bec33_array, label="ZO33")
plt.vlines(106, -1.0, -0.3, linestyle='dashed')
plt.ylim(-0.9,-0.38)
plt.ylabel("Z")
plt.xlabel("angle [° ]")
plt.legend()
```

```python
#===============================
# H原子のBEC対角成分
#===============================

bec11_list=[]
bec22_list=[]
bec33_list=[]

bec11_list2=[]
bec22_list2=[]
bec33_list2=[]

i_array=np.arange(80,130)

for i in range(80,130):
    print(i)
    # zero padding
    #count_zero = str(i).zfill(4)
    #print(count_zero)
    #BECarray[i,:,:,:]=np.load("WAT32_BEC/"+count_zero+".npy")
    ZH1=np.loadtxt("./water_anglechange/result/bec_H1_"+str(i)+".txt")
    bec11_list.append(ZH1[0][0])
    bec22_list.append(ZH1[1][1])
    bec33_list.append(ZH1[2][2])
    print(ZO)
    
    ZH2=np.loadtxt("./water_anglechange/result/bec_H2_"+str(i)+".txt")
    bec11_list2.append(ZH2[0][0])
    bec22_list2.append(ZH2[1][1])
    bec33_list2.append(ZH2[2][2])
    ##print(BECarray[i,:,:,:])
```

```python
# np.arrayに変換
bec11_array=np.array(bec11_list)
bec22_array=np.array(bec22_list)
bec33_array=np.array(bec33_list)

bec11_array2=np.array(bec11_list2)
bec22_array2=np.array(bec22_list2)
bec33_array2=np.array(bec33_list2)
```

```python
# 得られたdetたちを図示する．
import numpy as np
import matplotlib.pyplot as plt

# ここにグラフを描画する処理を記載する．
print(i_array)

# 縦線@平衡点
plt.vlines(106, -1.0, 0.5, linestyle='dashed')
plt.ylim(0.00, 0.5)

plt.plot(i_array,bec11_array, label="ZH1_11")
plt.plot(i_array,bec22_array, label="ZH1_22")
plt.plot(i_array,bec33_array, label="ZH1_33")

plt.plot(i_array,bec11_array2, label="ZH2_11")
plt.plot(i_array,bec22_array2, label="ZH2_22")
plt.plot(i_array,bec33_array2, label="ZH2_33")
plt.ylabel("Z")
plt.xlabel("angle [° ]")
plt.legend()
```

# DET(Z)の計算

## 得られたZを用いて，det(Z)を計算し，その変化を図示．

## 同時にZの対角成分のみの積の計算もやって，det(Z)とどれだけの違いが出るかをチェックしたい．

```python
#===============================
# O原子のdet(Z)
#===============================

# det(Z)のリスト
det_list=[]
det_list2=[]

i_array=np.arange(80,130)

for i in range(80,130):
    print(i)
    # zero padding
    #count_zero = str(i).zfill(4)
    #print(count_zero)
    #BECarray[i,:,:,:]=np.load("WAT32_BEC/"+count_zero+".npy")
    ZO=np.loadtxt("./water_anglechange/result/bec_O_"+str(i)+".txt")
    ## 対角成分の取り出し
    det_list.append(np.linalg.det(ZO))
    det_list2.append(ZO[0][0]*ZO[1][1]*ZO[2][2])
    #print(ZO)
    print(np.linalg.det(ZO))
    ##print(BECarray[i,:,:,:])
    
# np.arrayに変更
det_array=np.array(det_list)
det2_array=np.array(det_list2)

```

```python
#===============================
# データをファイルに保存
#===============================

test=np.block([[i_array], [det_array]])
print(test)
np.save("detZO", test)
```

```python
# 得られたdetたちを図示する．
import numpy as np
import matplotlib.pyplot as plt

# ここにグラフを描画する処理を記載する．
print(i_array)

plt.plot(i_array,det_array, label="det(ZO)")
plt.plot(i_array,det2_array, label="taikaku_ZO" ,linestyle='dashed')
plt.vlines(106, -1.0, 0.0, linestyle='dashed')
plt.ylim(-0.4,-0.10)
plt.ylabel("det(Z)")
plt.xlabel("angle [° ]")
plt.legend()
```

```python
#===============================
# H原子のBEC対角成分
#===============================

detH_list=[]
detH_list2=[]

taiH_list=[]
taiH_list2=[]


i_array=np.arange(80,130)

for i in range(80,130):
    print(i)
    # zero padding
    #count_zero = str(i).zfill(4)
    #print(count_zero)
    #BECarray[i,:,:,:]=np.load("WAT32_BEC/"+count_zero+".npy")
    ZH1=np.loadtxt("./water_anglechange/result/bec_H1_"+str(i)+".txt")
    detH_list.append(np.linalg.det(ZH1))
    taiH_list.append(ZH1[0][0]*ZH1[1][1]*ZH1[2][2])
    print(ZH1)
    
    ZH2=np.loadtxt("./water_anglechange/result/bec_H2_"+str(i)+".txt")
    detH_list2.append(np.linalg.det(ZH2))
    taiH_list2.append(ZH2[0][0]*ZH2[1][1]*ZH2[2][2])
    ##print(BECarray[i,:,:,:])
    
    
# np.arrayに変更
detH_array=np.array(detH_list)
detH_array2=np.array(detH_list2)

taiH_array=np.array(taiH_list)
taiH_array2=np.array(taiH_list2)
```

```python
#===============================
# データをファイルに保存
#===============================

#test=np.stack([i_array, detH_array], 1)
#test2=np.stack([test, detH_array2])
test2=np.block([[i_array], [detH_array], [detH_array2]])
print(test2)
np.save("detZH", test2)
```

```python
# 得られたdetたちを図示する．
import numpy as np
import matplotlib.pyplot as plt

# ここにグラフを描画する処理を記載する．
print(i_array)

# 縦線@平衡点
plt.vlines(106, -1.0, 0.5, linestyle='dashed')
plt.ylim(0.00, 0.025)

plt.plot(i_array,detH_array, label="det(ZH1)")
plt.plot(i_array,taiH_array, label="taikaku_ZH1",linestyle='dashed')

plt.plot(i_array,detH_array2, label="ZH2")
plt.plot(i_array,taiH_array2, label="taikaku_ZH2",linestyle='dashed')

plt.ylabel("Z")
plt.xlabel("angle [° ]")
plt.legend()
```
